```python
import requests
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
# define the endpoint
merk = "Audi"
merk_upper = merk.upper()
endpoint = f"https://opendata.rdw.nl/resource/m9d7-ebf2.json?merk={merk_upper}&$limit=150000"
endpoint
```

```python
response = requests.get(endpoint)
response
```

```python
# get the (json) data from the response
data = response.json()

# convert the json data (list[Dicts]) to a Pandas DataFrame
df_cars = pd.DataFrame(data)

# show the first 5 rows of the DataFrame
df_cars.head()
```

## Data Cleaning

* âœ… Only select needed columns
* Convert to the right data types
* Deal with empty values (NaN)
* Check for strange values


```python
# specify the required columns
columns_list = ['merk', 'handelsbenaming', 
                'datum_tenaamstelling', 'catalogusprijs',
                'aantal_cilinders', 'eerste_kleur']

df_cars_filtered = df_cars[columns_list]
df_cars_filtered.info()
```

```python
# conver for the right data types
df_cars_filtered['catalogusprijs'] = df_cars_filtered['catalogusprijs'].astype(float)
df_cars_filtered['aantal_cilinders'] = df_cars_filtered['aantal_cilinders'].fillna(0).astype(int)
df_cars_filtered['datum_tenaamstelling'] =  pd.to_datetime(df_cars_filtered['datum_tenaamstelling'],
               format="%Y%m%d")
```

```python
df_cars_filtered.info()
```

```python
# check for strange value
df_cars_filtered.describe()
```

```python
# check for the expensive car
df_cars_filtered.sort_values("catalogusprijs", ascending=False)
```

```python
# data augmentation
df_cars_filtered['jaar_maand_naam'] = df_cars_filtered['datum_tenaamstelling'].dt.strftime("%Y-%B")
df_cars_filtered['jaar_maand_nummer'] = df_cars_filtered['datum_tenaamstelling'].dt.strftime("%Y-%m")

df_cars_filtered.head()
```
